# PREDICTIONS avec algo KNN

```python
import pandas as pd
import joblib

# Charger le modèle et le scaler
model = joblib.load('billet_knn_model.pkl')
scaler = joblib.load('scaler.pkl')

# Charger le jeu de données de test
billets_test = pd.read_csv('billets_test.csv', sep=",", decimal='.')

# Préparation des données de test
features_test = billets_test[['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']]

# Normalisation des données de test
features_test_scaled = scaler.transform(features_test)

# Prédictions sur le jeu de données de test
predictions = model.predict(features_test_scaled)

# Obtenir les probabilités de classe
probabilities = model.predict_proba(features_test_scaled)

# Ajoutez les prédictions et les probabilités au DataFrame de test
billets_test['Predictions'] = predictions
billets_test['Probabilité_Classe_0'] = probabilities[:, 0]  # Probabilités pour la classe 0
billets_test['Probabilité_Classe_1'] = probabilities[:, 1]  # Probabilités pour la classe 1

# Affichage des résultats
display(billets_test)
```
